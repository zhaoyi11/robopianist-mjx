```python
import numpy as np
```

```python
A = np.random.randn(10, 3)
B = np.random.randn(10, 3)    
# calcualte the distance between keys and fingers
dist = np.full((10, 10), 100.)
for i, finger in enumerate(A):
    for j, key in enumerate(B):
        dist[i, j] = np.linalg.norm(key-finger)
```

```python
dist
```

```python
# Calculate the L1 distance
A_tiled = np.tile(A[:, np.newaxis, :], (1, 10, 1))  # Reshape A to (M, N, D)
B_tiled = np.tile(B[np.newaxis, :, :], (10, 1, 1))  # Reshape B to (M, N, D)
```

```python
# Calculate the absolute difference
abs_diff = np.abs(A_tiled - B_tiled)
```

```python
# Calculate the absolute difference
abs_diff = np.linalg.norm(A_tiled - B_tiled, axis=-1)

# # Sum along the last axis to get the L1 distance
# distances = abs_diff.sum(axis=2)
```
